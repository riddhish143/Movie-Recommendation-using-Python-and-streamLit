```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movies.head()
```

```python
# Join the both table on the basis of title

movies  = movies.merge(credits , on = 'title')
```

```python
movies.shape
```

```python
movies.head()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
    List = []
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List
```

```python
movies['genres'] = movies.genres.apply(convert)
```

```python
movies.genres.dtype
```

```python
movies.head()
```

```python
movies['keywords'] =movies.keywords.apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 6:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L 
```

```python
movies.cast = movies.cast.apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies.crew = movies.crew.apply(fetch_director)
```

```python
movies.head()
```

```python
movies.overview = movies.overview.apply(lambda X : X.split())
```

```python
movies.head()
```

```python
movies.genres = movies.genres.apply(lambda x : [i.replace(" " , "") for i in x])
```

```python
movies.keywords = movies.keywords.apply(lambda x : [i.replace(" " , "") for i in x])
movies.cast = movies.cast.apply(lambda x : [i.replace(" " , "") for i in x])
movies.crew = movies.crew.apply(lambda x : [i.replace(" " , "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['genres'] + movies['crew'] + movies['cast'] + movies['keywords'] + movies['overview']
```

```python
movies.head()
```

```python
movies_df = movies[['movie_id','title','tags']]
```

```python
movies_df
```

```python
import warnings
warnings.simplefilter('ignore')
```

```python
movies_df.tags = movies_df.tags.apply(lambda x :" ".join(x))
```

```python
movies_df.tags.dtype
```

```python
movies_df.tags = movies_df.tags.apply(lambda x :x.lower())
```

```python
movies_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies_df.tags = movies_df.tags.apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words = 'english' , max_features = 5000)
```

```python
vectors = cv.fit_transform(movies_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
movies_df[movies_df.title == "Avatar"].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse = True , key = lambda x : x[1])[1:10]
```

```python
def recommend(movie):
    index = movies_df[movies_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:10]:
        print(movies_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
